```python
data = pd.read_csv('hwe_encrypted.csv')
data
```

```python
hashed = []
for i in range(len(data)) : 
  #grab settings, product, ss strings, concat, hash
  tempSTR = str(data.iloc[i, 4]) + data.iloc[i, 5] + data.iloc[i,6]
  tempSTR = hl.md5(tempSTR.encode())
  tempSTR = tempSTR.hexdigest()
  #append to new array
  hashed.append(tempSTR)

#add new entries to dataframe
data['hashed'] = hashed
data = data.drop(['date', 'settings', 'product', 'subsystem'], axis = 1)
```
